## Identify outlier data 

```python
df1=df[7380::]
```

```python
df2=pd.DataFrame(df1,columns=['Team','Game','Home','Opponent','WINorLOSS','TeamPoints','OpponentPoints','TotalRebounds','Assists','Steals','Blocks','Turnovers','FieldGoals.','Opp.FieldGoals.'])
```

```python
for i in range(7380,9839):
    if df2.Home[i]=='Home':
        df2.Home[i]=1
        print(i)
    if df2.Home[i]=='Away':
        df2.Home[i]=0
        print(i)
 
            
```

```python
for i in range(7380,9839):
    if df2.WINorLOSS[i]=='L':
        df2.WINorLOSS[i]=0
        print(i)
    if df2.WINorLOSS[i]=='W':
        df2.WINorLOSS[i]=1
        print(i)
            
        
```

```python
df3=pd.DataFrame(df2,columns=['Home','WINorLOSS','TeamPoints','OpponentPoints','TotalRebounds','Assists','Steals','Blocks','Turnovers','FieldGoals.','Opp.FieldGoals.'])
```

```python
import numpy as np
df2.shape
```

```python
set1 = set()
for i in ['TeamPoints','OpponentPoints','TotalRebounds','Assists','Steals','Blocks','Turnovers','FieldGoals.','Opp.FieldGoals.']:
    mean=np.mean(df3[i])
    print(mean)
    stu=np.std(df3[i])
    print(stu)
    for j in range(7380,9830):
        if any([df3.loc[j,i]>(mean+2.42*stu)]):
            print('row  has been deleted',j)
            set1.add(j)
           # df3.drop([j],inplace=False)
            
        
```

```python
for item in set1:
    df3.drop([item],inplace=True)
```
